# Car Price Prediction

## Importing Relevent Packages and data

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from xgboost import XGBRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score,mean_squared_error
```

## Machine Learning and Preprcessing

```python
from sklearn.preprocessing import LabelEncoder, StandardScaler

# Encode categorical columns
categorical_cols = ['Car_Name', 'Fuel_Type', 'Selling_type', 'Transmission', 'fuel_type', 'Wheel']
df_encoded = df.copy()
le = LabelEncoder()

for col in categorical_cols:
    df_encoded[col] = le.fit_transform(df_encoded[col])

# Define features and target
X = df_encoded[['Car_Name', 'Year', 'Present_Price', 'Driven_kms', 'Fuel_Type',
                'Selling_type', 'Transmission', 'Owner', 'fuel_type', 'Wheel', 'engine']]
y = df_encoded['Selling_Price']

# Scale features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
x_train, x_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.3, random_state=24)

# XGBRegressor model
XG_model = XGBRegressor()
XG_model.fit(x_train, y_train)

# LinearRegression model
l_model = LinearRegression()
l_model.fit(x_train, y_train)

# RandomForestRegressor model
RF_model = XGBRegressor()
RF_model.fit(x_train, y_train)
```

```python
# Predictions from models
y_pred_xg = XG_model.predict(x_test)
y_pred_lr = l_model.predict(x_test)
y_pred_rf = RF_model.predict(x_test)

# R-squared Scores
r2_xg = r2_score(y_test, y_pred_xg) * 100
r2_lr = r2_score(y_test, y_pred_lr) * 100
r2_rf = r2_score(y_test, y_pred_rf) * 100

# Mean Squared Errors
mse_xg = mean_squared_error(y_test, y_pred_xg)
mse_lr = mean_squared_error(y_test, y_pred_lr)
mse_rf = mean_squared_error(y_test, y_pred_rf)

# Print Results with Bold Formatting
print(f"\033[1mXGBoost R-squared: {r2_xg:.2f}% | MSE: {mse_xg:.2f}\033[0m")
print(f"\033[1mLinear Regression R-squared: {r2_lr:.2f}% | MSE: {mse_lr:.2f}\033[0m")
print(f"\033[1mRandom Forest R-squared: {r2_rf:.2f}% | MSE: {mse_rf:.2f}\033[0m")
```
